```python
from qiskit.quantum_info import Operator,Statevector
from math import sqrt
# Initialize Standard Operators

Y = Operator([[0,-1j],[1j,0]])
H = Operator([[1/sqrt(2),1/sqrt(2)],[1/sqrt(2),-1/sqrt(2)]])
S = Operator([[1,0],[0,1j]])
T = Operator([[1,0],[0,1+1j]])

display(T.draw("latex"))

# Define Arbitrary StateVector

v = Statevector([1,0])

# Transform & Display StateVectors
display(v.draw("latex"))

v = v.evolve(H)
display(v.draw("latex"))

v = v.evolve(Y)
display(v.draw("latex"))

v = v.evolve(S)
display(v.draw("latex"))

v = v.evolve(T)
display(v.draw("latex"))
```
