```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector,Operator
from qiskit.visualization import plot_histogram

# Implement and Display Gates On A Single Qubit Circuit

qc = QuantumCircuit(1)
qc.h(0)
qc.s(0)
qc.h(0)
qc.y(0)
qc.t(0)

display(qc.draw(output="mpl"))
display(Operator.from_circuit(qc).draw("latex"))

# Circuit-Gate Conversion and It's Effect on a Standard Basis

ket0 = Statevector([1,0])
v = ket0.evolve(qc)
display(v.draw("latex"))

# Create Data-Sampling based on Circuit Implementation

stat = v.sample_counts(10000)
display(plot_histogram(stat))
```
