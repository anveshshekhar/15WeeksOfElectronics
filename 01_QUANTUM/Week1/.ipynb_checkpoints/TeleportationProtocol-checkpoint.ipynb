```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Statevector,Operator,random_statevector
from IPython.display import display,Latex

# Initialize Circuit Structure (Random Q, AB is Ï•+)
qr = QuantumRegister(3)
clx = ClassicalRegister(1)
cly = ClassicalRegister(1)
qc = QuantumCircuit(qr,clx,cly)
Q = random_statevector(2)
qc.initialize(Q,0)
qc.h(1)
qc.cx(1,2)
qc.barrier()
display(qc.draw(output="mpl"))

# 
```
