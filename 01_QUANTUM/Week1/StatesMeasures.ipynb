```python
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram
from numpy import sqrt

# Define Any State-Vectors

u = Statevector([1 / sqrt(2), 1 / sqrt(2)])
v = Statevector([(1 + 2.0j) / 3, -2 / 3])
w = Statevector([1 / 3, 2 / 3])

# Use Display to Draw/Print

display(u.draw("text"))
display(u.draw("latex"))
display(v.draw("latex"))

# Measure() Gives The Basis-Measure [Forces State-Vector to Chose a Basis Side - Based on The Probability {Amp^2}]
# Final State Upon Measurement Should be any Basis*(Phase Factor in Phase with Original State-Selection)
outcome, state = v.measure()
print(f"Measured: {outcome}\n Post-measurement state:")
display(state.draw("latex"))
display(v.draw("latex"))



# Statistics is a Tuple Giving Data of Any N-Measurements for Large-Dataset (or small, dependent on user input)
 
statistics = v.sample_counts(1000000)
plot_histogram(statistics)
```
