```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Latex,display

sampling_rate = 1000  
duration = 5.0        
N = sampling_rate * duration 

t = np.linspace(0, duration, int(N), endpoint=False) 

freq1, freq2 = 5, 7
signal = (1.5 * np.sin(2 * np.pi * freq1 * t) + 
          2.5 * np.sin(2 * np.pi * freq2 * t) +
          0.3 * np.random.normal(size=len(t)) )

fft_values = np.fft.fft(signal)

fft_freqs = np.fft.fftfreq(len(signal), d=1/sampling_rate)

half_n = len(signal) // 2
frequencies = fft_freqs[:half_n]
magnitudes = np.abs(fft_values)[:half_n] * (2.0 / len(signal))

cleanfft = fft_values.copy()
noises = (magnitudes < 0.2)[:half_n]
cleanfft[noises] = 0
recarr = np.fft.ifft(cleanfft)
recsignal = np
fig, ax, = plt.subplots(3, 1, figsize=(12, 8))

ax[0].plot(t[:500], signal[:500], color='b') 
ax[0].set_title('Time Domain Signal (First 0.5s)')
ax[0].set_xlabel('Time (s)')
ax[0].set_ylabel('Amplitude')
ax[0].grid(True)


ax[1].stem(frequencies, magnitudes, basefmt=" ") 
ax[1].set_title('Frequency Domain (FFT)')
ax[1].set_xlabel('Frequency (Hz)')
ax[1].set_ylabel('Magnitude')
ax[1].set_xlim(0, 10)  
ax[1].grid(True)

ax[2].plot(t[:500], recsignal[:500], color='b') 
ax[2].set_title('Error Corrected Signal (First 0.5s)')
ax[2].set_xlabel('Time (s)')
ax[2].set_ylabel('Amplitude')
ax[2].grid(True)
plt.tight_layout()
plt.show()
```
